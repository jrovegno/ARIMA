```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.tools.plotting import autocorrelation_plot

%matplotlib inline
```

```python
ruido_blanco = pd.DataFrame(pd.np.random.rand(1000))
```

```python
ruido_blanco.plot(title='Ejemplo Ruido Blanco', legend=False)
plt.xlabel('Tiempo')
plt.ylabel('ruido_blanco')
```

```python
autocorrelation_plot(ruido_blanco)
plt.ylim((-.2,.2))
```

```python
from statsmodels.stats.diagnostic import acorr_ljungbox
```

```python
lbvalue, pvalue = acorr_ljungbox(ruido_blanco)
plt.plot(lbvalue, pvalue)
```
