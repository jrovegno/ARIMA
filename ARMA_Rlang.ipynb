## Construyendo un modelo ARMA para series de tiempo

[Referencia](https://dlegorreta.wordpress.com/2015/04/27/construyendo-un-modelo-arma-para-series-de-tiempo)

```r
# Generamos ruido 
ruido_blanco = rnorm(1000, 0, 1)
```

```r
plot.ts(ruido_blanco, main="Ejemplo Ruido Blanco", xlab="Tiempo")
```

```r
# Graficamos correlograma
acf(ruido_blanco, main="Correlograma", col="2", lag=50, ylim = c(-.2, .2))
```

```r
# Calculamos la media
mean(ruido_blanco)
```

```r
#Prueba Ljung-Box

Box.test(ruido_blanco)
```
